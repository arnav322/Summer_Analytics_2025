# **Data Preprocessing**

## **Dropping some data**

```python
# Dropping SEQN column
train_df.drop('SEQN', axis=1, inplace=True)
test_df.drop('SEQN', axis=1, inplace=True)
```

```python
train_df.head()
```

```python
train_df.isnull().sum(), test_df.isnull().sum()
```

```python
# Dropping the rows with missing age_group columns.
train_df.dropna(subset=['age_group'], inplace=True)
```

```python
train_df.isnull().sum()
```

```python
train_df = train_df[train_df['PAQ605'] != 7]
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_df.drop('age_group', axis=1), train_df['age_group'], test_size=0.8, stratify=train_df['age_group'], random_state=42)
```

## **Handling Missing Data, Scaling and Encoding**

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(transformers = [
    ('num', Pipeline(steps=[
        ('imputer', IterativeImputer(random_state=42)),
        ('scaler', StandardScaler())
    ]), numerical_cols) ,

    ('cat', Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encode', OneHotEncoder(handle_unknown='ignore'))
    ]), categorical_cols) ,

], remainder='drop')

# Training Dataset
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Test Dataset
test_df_processed = preprocessor.transform(test_df)

# Encoding of target column in training dataset
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
X_train.shape, X_train_processed.shape, test_df.shape, test_df_processed.shape
```

# **Model Training**

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

lor = LogisticRegression(multi_class='multinomial', max_iter=1000, solver='lbfgs',penalty='l2', class_weight='balanced', random_state=42)
knn = KNeighborsClassifier(weights='distance' ,n_neighbors=1)
dt = DecisionTreeClassifier(random_state=42)

vote = VotingClassifier(estimators=[('lor', lor), ('knn', knn), ('dt', dt)],
                        voting='soft')

vote.fit(X_train_processed, y_train)
vote_pred = vote.predict(X_test_processed)

print("Voting Classifier:")
print(classification_report(y_test, vote_pred))

```

```python
vote_result = vote.predict(test_df_processed)
vote_file = pd.DataFrame(vote_result, columns=['age_group'])
vote_file.to_csv('Age_group_prediction.csv', index=False)
```
