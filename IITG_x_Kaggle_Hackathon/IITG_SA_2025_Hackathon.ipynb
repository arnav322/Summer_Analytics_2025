## **Filling Missing Values**

```python
ndvi_cols = df.drop(columns=['class']).columns
ndvi_cols
```

```python
for i in ndvi_cols:
  sns.distplot(df[i])
  plt.title(df[i])
  plt.show()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer, KNNImputer


knn = KNNImputer(n_neighbors=5)
df[ndvi_cols] = knn.fit_transform(df[ndvi_cols])



```

## **Handling Outliers**

```python
for i in ndvi_cols:
  sns.boxplot(df[i])
  plt.title(df[i])
  plt.show()
```

```python
df[ndvi_cols].iloc[0].plot(marker='o', title="Before Outlier Handling")
plt.xticks(rotation=90)
plt.show()
```

```python
for i in ndvi_cols:


  lower_limit = df[i].quantile(0.05)
  upper_limit = df[i].quantile(0.95)

  df[i] = np.where(df[i] < lower_limit, lower_limit,
                   np.where(df[i] >= upper_limit, upper_limit, df[i]))
```

```python
df[ndvi_cols].iloc[0].plot(marker='o', title="After Outlier Handling")
plt.xticks(rotation=90)
plt.show()
```

## **Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
```

```python
df['class'].value_counts()
```

## **Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop(columns= ['class']))
```

## **Training**

```python
# Split into features and target

#X = pd.DataFrame(X_scaled, columns=df.drop(columns= ['class']).columns)
#X = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])






X = df.drop(columns= ['class'])
y = df['class']

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Fit multinomial logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    #class_weight='balanced',
    penalty='l2',
    max_iter=1000
)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Classification report with all original class labels
print(classification_report(
    y_test,
    y_pred,

))
```
